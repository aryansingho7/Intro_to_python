```python
#py program to illustarte itrating over a list
print('list iterating')
l=('geeks','for','geeks')
for i in l:
    print(i)

#itertaing over tuples
print('tuplr iterating')
t=('geeks', 'for','geeks')
for x in t:
    print(x)

#iterating over strings
print('string iteration')
s="geeks"
for i in s:
    print(i)

 #iteration over dictionary
print('dictionary iteration')
d=dict()
d["xyz"]=123
d["abc"]=345
for i in d:
    print('%s %d' %(i,d[i]))
```

```python
fruits=['apple','oranges','kiwi']

#creating an iterator object
#from that iterable i.e fruits
iter_ojb= iter(fruits)
#infinite loop
while True:
    try:
        #getting the next item 
        fruits= next(iter_ojb)
        print(fruits)
    except StopIteration:
        #if Stopiteration is raised 
        #break from loop
        break
```
