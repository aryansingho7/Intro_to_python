```python
#python program to show range() basics

#print a number
for i in range(10):
    print(i, end=' ')
print()

#using range for iteration
l=[10,20,30,40]
for i in range(len(l)):
    print(l[i], end=' ')
print()

#performing sum of neutral number
sum = 0
for i in range(1, 11):
    sum+= i
print('sum of first 10 natural numbers:', sum)
```

```python
#py program to print whole number using range()
#range(stop)- one argument
#printing first 10 whole numbers
for i in range(10):
    print(i, end=" ")
print()

#printing first 20 whole numbers
for i in range(20):
    print(i, end=' ')
print()
```

```python
#range(start, stop)- two arguments
#py program to pritn natural numbers using range

#printing thevnatural number upto 20
for i in range(1,20):
    print(i, end=' ')
print()


#printing a natural number form 5 to 20
for i in range(5,20):
    print(i, end=" ")
print()
```

```python
#range( start, stop step)- third type
#python program to print all number divisible by 3 

#using range to orint numbers
#divisible by 3
for i in range(0,30,3):
    print(i, end=' ')
print()

#using the range to print number divisible by 5
for i in range(0,30,5):
    print(i, end=" ")
print()
```

```python
#incrementing

#py program to increment with range()

#increment by 2
for i in range(2,25,2):
    print(i, end=' ')
print()

#increment by 4
for i in range(0,30,4):
    print(i, end=" ")
print()

#increment by 3
for i in range(15, 25,3):
    print(i, end=' ')
print()
```

```python
#decrement 

#py program to decrement with range()

#increment by -2
for i in range(25,2,-2):
    print(i, end=" ")
print()

#increment by -4
for i in range(30,1,-4):
    print(i, end=" ")
print()

#incremnet by -3
for i in range(25,-6,-3):
    print(i, end=" ")
print()
```

```python
#CONCATENATION OF TWO RANGES() FUCN
#using chain()

from itertools import chain

#using chain methord
print('concatenation the result')
res= chain(range(5), range(10,20,2))

for i in res:
    print(i, end=" ")
```

```python
#accessing range()
#py program to demostrate range function

ele= range(10)[0]
print('first element:', ele)

ele = range(10)[-1]
print('\n last elemnt:', ele)

ele = range(10)[4]
print('\n fifth element:', ele)
```
