```python
#python program to illustrate 
#while and wile-else loop
i=0
while(i<3):
    i=i+1
    print('hello geek')
    
#checks is the list is still 
#contains any element
a=[1,2,3,4]
while a:
    print(a.pop())

i=10
while i<12:
    i+=1
    print(i)
    break
else: #not executed as there is a break 
    print('no break')
```

```python
#python program to illusatrate
#iteration over alist
print('list iteration')
l = ['geeks','for','geeks']
for i in l:
    print(i)

#iterating over a tuple(immuatble)
print('tuple iteration\n')
t=('geeks','for','geeks')
for i in t:
    print(i)

#iterating over a string
print('string iteration')
s='geeks'
for i in s:
    print(i)

#iteration over dictionary
print('dictionary iteration')
d =dict()
d['xyz']=123
d['abc']=345
for i in d:
    print('%s %d' %(i,d[i]))
```
