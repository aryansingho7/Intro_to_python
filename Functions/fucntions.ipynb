```python
#A simple python fucntion to check 
#whether s is even or odd

def evenodd(x):
    if (x % 2==0):
        print("even")
    else:
        print('odd')


#driver code
evenodd(4)
evenodd(99)
```

```python
#return the sqaure value of the entered number

def square_value(num):
    return num**2

print(square_value(2))

print(square_value(818))

print(square_value(-4))
```

```python
#here x is the new reference to same list 1st
def myfun(x):
    x[0]=20

#driver code, 1st will be modified after calling ones
lst=[10,11,12,13,14,15]
myfun(lst)
print(lst)
```
