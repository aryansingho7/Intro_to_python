```python
#python program to illustrate 
#*args for variable number of arguments
def myfun(*argv):
    for arg in argv:
        print(arg)
myfun('hello', 'welcome', 'to', 'geeksforgeeks')
```

```python
#python program to illustrate 
#*args with first extra argument

def myfun(arg1, *argv):
    print('first argument:', arg1)
    for arg in argv:
        print('next arguemnt through *argv:', arg)
myfun('hello', 'welcome', 'to', 'geeksforgeeks')
```

```python
#**kwargs
#python program to illustrate *kargs for variable number of keyword arguments

def myfun(**kwargs):
    for key, value in kwargs.items():
        print("%s==%s" %(key, value))

#drivers codee
myfun(first='geeks', mid='for', last='geeks')
```

```python
#using *args and **kwargs to call function

#EXAMPLE

def myfun(arg1, arg2, arg3):
    print('arg1:', arg1)
    print('arg2:', arg2)
    print('arg3', arg3)

#now we can use *args or **kwargs to pass arguments to this function:
args=('geeks', 'for', 'geeks')
myfun(*args)

kwargs= {'arg1':'geeks', 'arg2':'for', 'arg3':'geeks'}
myfun(**kwargs)
```

```python
#using *args and **kwargs in the same line call a function

def myfun(*args,**kwargs):
    print('args:', args)
    print('kwargs:', kwargs)

#now we can use both *args, **kwargs
#to pass arguments to this fucntion

myfun('geeks','for','geeks',first='geeks',mid='geeks',last='geeks')
```
